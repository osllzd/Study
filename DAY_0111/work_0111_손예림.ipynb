(1) 모듈 로딩------------------------------------------------------------------

```python
from pandas import DataFrame,read_csv
```

(2) 데이터 준비 --------------------------------------------------------------

```python
dataFile='../DATA/movies.csv'
```

(3) 데이터 저장 ---------------------------------------------------

```python
movieDF=read_csv(dataFile)
```

```python
movieDF
```

(4) DataFrame의 속성들 출력

```python
DF=movieDF.columns
DF
```

(5) DataFrame의 컬럼별로 따로 저장하기

```python
cols={}
for s in movieDF.columns:
    cols[s] = movieDF[s]
# cols
print(type(cols["Rank"]))
print(cols["Rank"])
```

(6) 컬럼별 데이터들의 속성 즉, index, values, shape, ndim, dtype정보 출력

```python

for colName in movieDF.columns:
    print(f"--------------------------------------------------{colName}의 속성 정보--------------------------------------------------")
    print(f"index : {movieDF[colName].index}")
    print(f"vaules : {movieDF[colName].values}")
    print(f"shape : {movieDF[colName].shape}")
    print(f"ndim : {movieDF[colName].ndim}")
    print(f"size : {movieDF[colName].size}")
    print(f"dtype : {movieDF[colName].dtype}")
    print()
```

(7) 컬럼별 인덱스를 변경 

```python
movieDF.columns=['A','B','C','D','F']
movieDF
```

(8) 컬럼별 데이터 즉 값 변경

```python
movieDF.iloc[0,0]='Happy new year 2024!'
movieDF.iloc[0,1]='Merry christmas~~'
movieDF
```

(9) DataFrame의 원소들 모두 출력 단 2가지 방법 사용

```python
for row in movieDF.index:
    print(row)
    for col in movieDF.columns:
        print('----',col,movieDF.loc[row,col])
```

```python
for row in range(movieDF.shape[0]):
    print(row)
    for col in range(movieDF.shape[1]):
        print(row,col,movieDF.iloc[row,col])
```
