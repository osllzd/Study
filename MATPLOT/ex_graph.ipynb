## [데이터 시각화 맛보기 ]

(1) 모듈 로딩 <hr>

```python
import matplotlib.pyplot as plt               # 그래프 관련 모듈
import matplotlib as mpl               # 패키지 관련
import matplotlib.font_manager as fm   # 패키지 리소스 즉 자원 중 폰트 관련 모듈

import pandas as pd
```

(2) 데이터 준비 <hr>

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = list(range(1,100,10)) # 0~ 99
y = [d**2 for d in x] #for d in x: y.append(d**2)
x,y
```

(3) 그래프 그리기 <hr>

- [기본] 라인 / 선 그래프
    * 시간의 변화를 표현해주는 그래프
    * 연속적/순서 있는 데이터를 시각화할 경우 주로 사용됨

```python
# plot(data) => y축 데이터로 사용
# plot(data1, data2) => x축에 data1, y축에 data2 데이터로 사용
# 반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환
ret=plt.plot(x,y)
type(ret), len(ret), type(ret[0])

# 파이썬에서 표준출력 기능 내장 함수 => print() 동일 역할
plt.show()
ret[0].get_xdata()
```

### x,y 축 레이블 설정 및 눈금 설정  
- x,y 축 데이터에 대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태 설정 => plt.xticks(), plt.yticks(), plt.tick_param()

```python
# 그래프 그리기
plt.figure(facecolor='yellow',figsize=(15,10))
plt.plot(y,'ro--', y2,'g^')
plt.title("[테스트]")

# x, y 축 레이블 
plt.xlabel("값",loc='right',fontsize='xx-large',color='red')
plt.ylabel("결과",loc='top',fontsize='xx-large',color='blue',rotation='horizontal')

# x, y축 눈금 설정
plt.xticks(ticks=[0, 2, 4, 6, 8], labels=['A', 'B', 'C', 'D', 'E'], color='red')
plt.yticks(color='blue')
```

```python
# 그래프 그리기
plt.plot(y,'ro--', y2,'g^')
plt.title("[테스트]")

# x, y 축 레이블 
plt.xlabel("값",loc='right',fontsize='xx-large',color='red')
plt.ylabel("결과",loc='top',fontsize='xx-large',color='blue',rotation='horizontal')

# x, y축 눈금 설정
plt.xticks(ticks=[0, 2, 4, 6, 8], labels=['A', 'B', 'C', 'D', 'E'], color='red')
plt.yticks(color='blue')

#범례 설정
plt.legend(labels=['red결과','green결과'],loc=10,edgecolor='green')

# 그래프 안에 텍스트 출력
plt.text(5,1000,'Merry Christmas')
```

## 텍스트 추가  
- 그래프 내에 위치 지정후 택스트 추가 설정 => plt.text(시작x, 시작y, 출력텍스트)

```python
# 그래프 그리기
plt.plot(y,'ro--', y2,'g^')
plt.title("[테스트]")

# x, y 축 레이블 
plt.xlabel("값",loc='right',fontsize='xx-large',color='red')
plt.ylabel("결과",loc='top',fontsize='xx-large',color='blue',rotation='horizontal')

# x, y축 눈금 설정
plt.xticks(ticks=[0, 2, 4, 6, 8], labels=['A', 'B', 'C', 'D', 'E'], color='red')
plt.yticks(color='blue')

#범례 설정
plt.legend(labels=['red결과','green결과'],loc=10,edgecolor='green')

# 그래프 안에 텍스트 출력
plt.text(2,2000,'Merry Christmas',backgroundcolor='yellow',color='red')

box = {
    'facecolor':'.75',
    'edgecolor':'k',
    'boxstyle':'round',
'alpha':0.2}
plt.text(2, 2000, "Merry Christmas",bbox=box)
```
