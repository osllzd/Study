## [파이 pie plot ] 

- 전체 데이터에서 특정 요소의 비중을 시각화하는 그래프


(1) 모듈 로딩 <hr>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터
labels = 'Frogs' , 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

# 1개의 subplot을 ax에 저장해서 반환
fig, ax=plt.subplots()
ax.pie(sizes, labels=labels)
plt.show()
```

```python
# 1개의 subplot을 ax에 저장해서 반환
fig, ax=plt.subplots()

ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```
