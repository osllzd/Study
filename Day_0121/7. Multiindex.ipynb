- MultiIndex 객체

```python
import pandas as pd
```

```python
address=("8809 Flair Square", "Toddside", "IL", "37206")
address
```

```python
addresses = [("8809 Flair Square", "Toddside", "IL", "37206"),
             ("9901 Austin Street", "Toddside", "IL", "37206"),
             ("905 Hogan Quater", "Franklin", "IL", "37206"),]
addresses            
```

```python
pd.MultiIndex.from_tuples(addresses)
pd.MultiIndex.from_tuples(tuples=addresses)
```

```python
row_index = pd.MultiIndex.from_tuples(tuples=addresses, names= ["Street", "City", "State", "Zip"])
row_index
```

- MultiIndex 생성 => DataFrame 연결

```python
data = [["A", "B+"], ["C+", "C"],["D-","A"]]
columns = ["Schools", "Cost of Living"]
area_grades=pd.DataFrame(data=data, index=row_index, columns=columns)
area_grades
```

```python
column_index=pd.MultiIndex.from_tuples([("culture", "resturants"), 
                                        ("culture","museums"),
                                        ("services","police"),
                                        ("services","schools"),
                                       ])
```

```python
column_index
```

```python
data = [["C-","B+","B-","A"],["D+","C","A","C+"],["A-","A","D","F"],]
```

```python
pd.DataFrame(data=data, index=row_index, columns=column_index)
```
