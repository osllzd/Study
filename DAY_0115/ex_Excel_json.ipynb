## [Json ==> DataFrame 저장]

(1) 데이터 준비

```python
file3='../DATA/nobel.json'
```

(2) Json ==> DataFrame으로 저장

```python
novelDF=pd.read_json(file3)
novelDF.info()
```

(3) 데이터 확인

```python
print(novelDF.head(3), novelDF.tail(3), sep='\n\n')
```

(4) 데이터 추출

```python
# 행 추출 => iloc[] or loc[]
type(novelDF.iloc[0]), novelDF.iloc[0]
```

```python
novelDF.iloc[0]['prizes']
```

```python
# 행 추출 => iloc[[n1,...]]/loc[]
novelDF.iloc[[0,3]]
```

```python
oneData = novelDF.iloc[0]['prizes']
type(oneData)
```

```python
oneData.keys(), oneData.items()
```

```python
for k,v in oneData.items():
    print(f'[{k}]\n===>{v}')
    if k == 'laureates':
        for d in v: #v => dict원소로 구성된 list
            for k1, v1 in d.items():
                print('-', k1, v1)
```

-json 중첩 데이터의 다중 반복문 처리 ==> 단일 리스트 형태: 평탄화/ 정규화

```python
pd.json_normalize(oneData)
```

```python
pd.json_normalize(oneData,record_path='laureates')
```

```python
pd.json_normalize(oneData,record_path='laureates', meta=['year', 'category'])
```

- 모든 행에 대한 정규화/평탄화 진행

```python
# 컬럼 읽기 => DF객체변수명[컬럼명], DF객체변수명.컬럼명
# ====> 모든행에 record_path에 해당하는 키가 존재해야 함!!
# === > 없는 행에 record_path에 해당하는 키 추가
for data in novelDF.prizes:
    print(type(data),data.keys(), end='\n\n')
    # 해당 dict에 laureates 키가 없다면 추가
    data.setdefault('laureates',[])
```

```python
novelDF2=pd.json_normalize(novelDF.prizes,record_path='laureates', meta=['year','category'])
novelDF2.info()
```

- (7) Json의 다양한 방식의 DataFrame 저장하기 => orient 파라미터

```python
# 데이터 => 키 : {값}
json_data='''
[
{"id" : "1","name":"sravan","age":22},
{"id" : "2","name":"harsha","age":22},
{"id" : "3","name":"deepika","age":21},
{"id" : "4","name":"jyothika","age":23}
]
'''
```

- orient = 'record' 설정 => Json파일 데이터를 행 단위로 읽어서 DataFrame에 저장

```python
# Convert Json string with records orient to a DataFrame
df=pd.read_json(json_data, orient='records')
# Display the Datafrome
print(df)
```

```python
# 데이터 => 키 : {값}
json_data2='''
{
"student-1":{"id" : "1","name":"sravan","age":22}, 
"student-2":{"id" : "2","name":"harsha","age":22},
"student-3":{"id" : "3","name":"deepika","age":21},
"student-4":{"id" : "4","name":"jyothika","age":23}
}
'''
```

```python
# 키 -> 컬럼
df2=pd.read_json(json_data2).T #, orient='columns')
print(df2)
```

```python
#키 -> 인덱스로 설정
df3 = pd.read_json(json_data2, orient='index')
print(df3)
```

```python
# JSON Format 
json_data4 = '''
[
    [ "1", "sravan",22 ],
    [ "2", "harsha",22 ],
    [ "3", "deepika",21 ],
    [ "4", "jyothika",23 ]
]
'''
```

```python
df4 = pd.read_json( json_data4, orient='values')
print(df4)
```
