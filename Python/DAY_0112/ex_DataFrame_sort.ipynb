## (1) 모듈 로딩

## (2) 데이터 준비

```python
datafile='../DATA/movies.csv'
```

## (3) 데이터 저장 => DataFrame

## (4) 데이터 확인

    * 데이터 확인 관련 메서드

```python
# 데이터의 전체적인 구조 및 데이터 종류 요약
moviesDF.info()
```

```python
# 컬럼별 통계적인 데이터 정보 제공
twoDF=moviesDF.describe(include='all')
twoDF
```

```python
# 수치 컬럼의 최소, 최대, 평균 데이터 추출
```

```python
twoDF.loc[['min','max','mean']]
```

- ## 행 => 열, 열 => 행 치환 기능 속성 및 메서드

```python
pd.DataFrame([twoDF.Rank.values,twoDF.Year.values],
            columns=twoDF.index,index=twoDF.columns)
```

- 속성 : DF객체변수명.T => 원본 데이터 변경 안됨

```python
twoDF.T
```

- 메서드: DF객체변수명.transpose()

## (5) 데이터 정렬
    - 인덱스 기반 정렬 => 행인덱스/열인덱스


```python
# 행 인덱스 기반 정렬 => DF.sort_index()Z
moviesDF.sort_index(ascending=False,axis=0)
```

```python
# 컬럼 인덱스 기반 정렬 
moviesDF.sort_index(axis='columns',ascending=False)
```

```python
moviesDF.sort_values(by=['Year','Title'])
```

-[실습] 다양한 정렬 <hr>

```python
from math import nan

data={'col1':['A','A','B', nan, 'D', 'C'],
      'col2':[2, 1, 9, 8, 7, 4],
      'col3':[0, 1, 9, 4, 2, 3],
      'col4':['a', 'b', 'c', 'D', 'e', 'F']}
```

```python
dataDF = pd.DataFrame(data)
dataDF
```

- col1칼럼을 기준으로 오름차순 정렬

```python
# NaN 즉, 반환 데이터(결측치)값의 위치: (기) 제일 마지막
dataDF.sort_values(by='col1',na_position='first')
```

```python
#col4 컬럼을 기준으로 오름차순 정렬
dataDF.sort_values(by='col4')
```

```python
dataDF.sort_values(by='col4', key=lambda col:col.str.upper())
```
