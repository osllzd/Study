### [병합 - (2) merge ]
- 두 개의 DF에서 특정 컬럼을 기준으로 데이터를 합치는 것
- 두 DFd에 합치는 기준이 되는 컬럼 지정 필요
- 지정된 컬럼명이 없으면 동일한 컬럼명을 기준으로 합쳐짐

- (2-1) 데이터 준비 <hr>

```python
f1='../DATA/stock price.xlsx'
f2='../DATA/stock valuation.xlsx'
```

- (2-2) 데이터 저장 <hr>

```python
df1=pd.read_excel(f1)
df2=pd.read_excel(f2)
```

- (2-3) 데이터 확인 <hr>

```python
df1.info()
```

```python
df2.info()
```

- (2-4) DataFrame 병합 <hr>

- (2-4-1) merge()

```python
# [기본] 두 개의 DF에서 동일 컬럼명을 찾아서 두 DF에 존재하는 컬럼명의 행 데이터만 합치기
pd.merge(df1,df2)
```

```python
df1.shape, df2.shape
```

```python
# [설정 => how 'outer'] 두 개의 DF에서 동일 컬럼명을 찾아서 두 DF에 존재하는 컬럼명의 행 데이터만 합치기
pd.merge(df1,df2, how='outer')
```

```python
# [설정 => how 'left'/'right'] 두 개의 DF에서 동일 컬럼명을 찾아서 왼쪽/오른쪽 DF데이터를 기준으로 병합
pd.merge(df1,df2, how='left')
```

```python
pd.merge(df1,df2, how='right')
```

```python
# [설정 => how 'cross'] 두 개의 DF에서 동일 컬럼명을 찾아서 왼쪽/오른쪽 DF데이터를 기준으로 병합
pd.merge(df1,df2, how='cross')
```
